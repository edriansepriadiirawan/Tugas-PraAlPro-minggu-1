```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
x = np.linspace(0,10)
y = np.sin(x)
z = np.cos(x)
plt.plot(x,y,'b',x,z,'r')
plt.xlabel('radiants')
plt.ylabel('value')
plt.title('plotting demonstration')
plt.legend(['sin','cos'])
plt.grid()
```

```python
# Definisikan variabel
berat_emas_awal = 25
harga_emas_awal = 650000
harga_emas_sekarang = 685000
berat_emas_baru = 15
harga_emas_baru = 715000

# Hitung keuntungan Gerard dalam Scenario 1
keuntungan_rp1 = (harga_emas_sekarang - harga_emas_awal) * berat_emas_awal
keuntungan_persen1 = (keuntungan_rp1 / (harga_emas_awal * berat_emas_awal)) * 100
persen1 = int(round(keuntungan_persen1))

# Hitung keuntungan Gerard dalam Scenario 2
total_berat_emas = berat_emas_awal + berat_emas_baru
total_harga = berat_emas_awal * harga_emas_awal + berat_emas_baru * harga_emas_sekarang
keuntungan_rp2 = total_berat_emas * harga_emas_baru - total_harga
keuntungan_persen2 = (keuntungan_rp2 / total_harga) * 100
persen2 = int(round(keuntungan_persen2))

# Tampilkan hasil
print("Hasil Scenario 1:")
print("Berapa keuntungan yang didapat oleh Gerard (dalam Rp)?:", keuntungan_rp1)
print("Berapa keuntungan yang didapat oleh Gerard (dalam %)?:", persen1, "%")
print("\nHasil Scenario 2:")
print("Berapa keuntungan yang didapat oleh Gerard (dalam Rp)?:", keuntungan_rp2)
print("Berapa keuntungan yang didapat oleh Gerard (dalam %)?:", persen2,"%")
```
