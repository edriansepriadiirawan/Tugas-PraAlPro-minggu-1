```python
import re
from datetime import datetime

def proses_tanggal(teks):
    pola = r'\b\d{4}-\d{2}-\d{2}\b'
    tanggal_ditemukan = re.findall(pola, teks)
    hasil = []
    sekarang = datetime.now()

    for tanggal in tanggal_ditemukan:
        tanggal1 = datetime.strptime(tanggal, '%Y-%m-%d')
        tanggal_baru = tanggal1.strftime('%d-%m-%Y')
        selisih_hari = (sekarang - tanggal1).days
        hasil.append(f"{tanggal} selisih {selisih_hari} hari")

    return teks, hasil

teks = """
Pada tanggal 1945-08-17 Indonesia merdeka. Indonesia memiliki beberapa pahlawan nasional,
seperti Pangeran Diponegoro (TL: 1785-11-11), Pattimura (TL: 1783-06-08) dan Ki Hajar Dewantara (1889-05-02).
"""

isi, hasil = proses_tanggal(teks)

print("contoh isi teks:")
print(isi)

print("\nHasil Pemrosesan Tanggal:")
for hasil in hasil:
    print(hasil)
```

```python
import re
import random
import string

def generate_password():
    huruf = string.ascii_letters
    angka = string.digits
    password = ''.join(random.choices(huruf, k=3))
    password += ''.join(random.choices(angka, k=2))
    password += ''.join(random.choices(huruf + angka, k=3))
    return password

emails = [
    "anton@mail.com dimiliki oleh anton",
    "budi@gmail.co.id dimiliki oleh budi",
    "slamet@getnada.com dimiliki oleh slamet",
    "matahari@tokopedia.com dimiliki oleh matahari"
]

for email in emails:
    match = re.search(r'(\S+@\S+)\s+dimiliki oleh\s+(\S+)', email)
    if match:
        alamat_email = match.group(1)
        nama_pengguna = match.group(2)

        kata_sandi = generate_password()

        print(f"{alamat_email} username: {nama_pengguna} , password: {kata_sandi}")
```
